Продажи по товарам

```python
groupped_df = df.groupby('DR_NDrugs').agg({'DR_Kol': sum})
groupped_df
```

Вклад каждого товара

```python
groupped_df = (groupped_df['DR_Kol'] / sum(groupped_df['DR_Kol'])).sort_values(ascending=False)
groupped_df
```

Сумма продаж накопительным итогом

```python
groupped_df = groupped_df.cumsum().reset_index()
groupped_df
```

Определение групп товаров

```python
import numpy as np
```

```python
groupped_df['ABC'] = np.where(groupped_df['DR_Kol'] < 0.8, 'A', np.where(groupped_df['DR_Kol'] < 0.95, 'B', 'C'))
```

<b>Функция ABC-анализа</b>

```python
def perform_ABC(df, index):
    # На вход подается датафрейм Pandas (df) и столбец, по которому будет проводиться группировка (index)

    cols = list(df.columns)
    cols.remove(index)
    ABC = df.groupby(index).agg({col: sum for col in cols})
    for col in cols:
        ABC[f'rel_{col}'] = (ABC[col] / sum(ABC[col]))
        ABC = ABC.sort_values(f'rel_{col}', ascending=False)
        ABC[f'cumsum_{col}'] = ABC[f'rel_{col}'].cumsum()
        ABC[f'ABC_{col}'] = np.where(ABC[f'cumsum_{col}'] < 0.8, 'A', np.where(ABC[f'cumsum_{col}'] < 0.95, 'B', 'C'))
    return ABC[cols + [f'ABC_{col}' for col in cols]]
```

<b>Динамика продаж</b>

```python
df_new = pd.read_csv('data/data1.csv', encoding='1251')
```

Сумма продаж за каждый день

```python
df_new = df_new.groupby('DR_Dat').apply(lambda x: pd.Series({'revenue': sum(x['DR_Kol'] * x['DR_CRoz'] - x['DR_SDisc'])})).reset_index()
df_new
```

Динамика с помощью функции rolling

```python
df_new['revenue_d'] = df_new['revenue'].rolling(2).apply(lambda x: (x.iloc[1] - x.iloc[0]) / x.iloc[0])
df_new
```

Динамика с помощью функции shift

```python
df_new['revenue_shifted'] = df_new['revenue'].shift(1)
df_new
```

```python
df_new['revenue_d2'] = (df_new['revenue'] - df_new['revenue_shifted']) / df_new['revenue_shifted']
df_new
```

Динамика с помощью функции pct_change

```python
df_new['revenue_d3'] = df_new['revenue'].pct_change(1)
df_new
```

<b>XYZ-анализ</b>

Продажи по дням с разбивкой по товарам

```python
df_new2 = pd.read_csv('data/data1.csv', encoding='1251')
```

```python
df_new2 = df_new2.groupby(['DR_Dat', 'DR_NDrugs']).agg({'DR_Kol': sum})
df_new2
```

Отбор только тех товаров, что продавались несколько дней

```python
df_n2 = df_new2.reset_index().groupby('DR_NDrugs').agg({'DR_Dat': 'count'}).reset_index()
xyz_names = list(df_n2[df_n2['DR_Dat'] > 1]['DR_NDrugs'])
xyz_names
```

```python
df_new2 = df_new2.reset_index()
df_new3 = df_new2[['DR_NDrugs', 'DR_Kol']]
df_new3
```

Определение коэффициента вариации

```python
df_new3 = df_new3[df_new3['DR_NDrugs'].isin(xyz_names)]
df_new3 = df_new3.groupby('DR_NDrugs').apply(lambda x: x.std()/x.mean())
df_new3
```

Определение групп XYZ

```python
df_new3['XYZ'] = np.where(df_new3['DR_Kol'] < 0.1, 'X', np.where(df_new3['DR_Kol'] < 0.25, 'Y', 'Z'))
df_new3
```
