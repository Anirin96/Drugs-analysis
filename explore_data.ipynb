Связь между объемом продаж и розничной ценой с разбивкой по датам

```python
import numpy as np
```

```python
df.groupby(['dt', 'drug'])[['roz', 'kol']].agg({
    'roz': np.mean,
    'kol': sum
})
```

```python
df['rroz'] = df.roz.apply(lambda x: round(x, -2))
```

```python
c = df.groupby(['dt', 'rroz'])['kol'].agg(sum).reset_index()
```

```python
c = c[c.loc[:, 'rroz'] < 3000]
```

```python
sb.scatterplot(x=c.rroz, y=c.kol, hue=c.dt)
```

```python
sb.jointplot(x=c.rroz, y=c.kol, hue=c.dt)
```
